```python
import pandas as pd
data = pd.read_csv(p, sep="\t")
```

```python
import numpy
from bert_serving.client import BertClient
bc = BertClient(check_length=False)
```

```python
data
```

```python
x1=[]
x2=[]
x3=[]
y=[]
labels={}
i=0
for ind, row in data.iterrows():
    if row['label'] not in labels:
        labels[row['label']]=i
        i+=1
    y.append(labels[row['label']])
    x1.append(row['turn1'])
    x2.append(row['turn2'])
    x3.append(row['turn3'])
    
```

```python
x1_vec=bc.encode(x1)
x2_vec=bc.encode(x2)
x3_vec=bc.encode(x3)
```

```python
len(x1_vec)
```

```python
import sklearn as sk 
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=0).fit(x1_vec[0:27000], y[0:27000]) 
```

```python
LR.predict(x1_vec[27000:])  
round(LR.score(x1_vec[27000:],y[27000:]), 4)  
```
